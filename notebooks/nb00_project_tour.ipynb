# Project Tour ? Revenue Cycle Decision Pack (Synthetic Claims)

## What this is
A decision-ready analytics workflow that mirrors an exec ops report:
**Overview ? Drivers ? Workqueue**, using marts-only outputs from dbt.

## How to read (90 seconds)
1) **NB-03 Exec Overview**: what changed, what?s stable, what to watch
2) **NB-04 Drivers**: which segments contribute most to the movement (not causality)
3) **NB-05 Workqueue**: prioritized queue preview (proxy-based, not guaranteed recovery)

- NB-03: docs/executive_summary.md + docs/decision_memo_latest_complete_week.md
- NB-04: docs/driver_pareto_memo_latest_period.md
- NB-05: docs/workqueue_memo_latest_week.md
